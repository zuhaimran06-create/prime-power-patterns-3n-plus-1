```python
f=tan(1/x)/x**2
int=integrate(f,(x,1,oo))
print('The improper integral converges to',int.evalf())
```

e) use remainder theorem to find how many terms needed to sum within 0.001

```python
x=symbols('x',real=True)
N=symbols('N',positive=True,real=True)
f=tan(1/x)/x**2

errbound=integrate(f,(x,N,oo))
errbound
```

```python
Eq=errbound-0.001
solve(Eq,N)
```

```python
print('The minimum number of terms is at least 23')
```

findthe 23rd partial sum

```python
a1_23=[bn.subs(n,i).evalf() for i in range(1,24)]
S23=sum(a1_23)
print('S23=',S23.evalf())
```

Fidn the actual sum

```python
from mpmath import nsum,inf
S=nsum(lambda x:tan(1/x)/x**2,[1,inf])
print('The actual sum is',S)
```

```python
print('The error is',abs(S-S23))
```
